```python
M_arr = []
g = Blotto(3, (5, 5))

t_arr = np.array([2**i for i in range(3, 13)])
for t in t_arr:
    SE = Single_Evaluation(g, t, agent=Regret_Minimisation_Agent)
    strat=SE.get_strategies()[0]
    
    M = evaluate(Agent(g, 0, strat),round=1000)
    M_arr.append(M)
```

```python
plt.plot(t_arr, M_arr, label="Worst payoff")
plt.plot(t_arr, 1/np.sqrt(t_arr), label="1/sqrt(x)")
plt.xlabel("Number of rounds played")
plt.legend()
plt.savefig("./plots/simple game/worst_performance")
```
